```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Carrega o shapefile dos municípios brasileiros
brasil_municipios = gpd.read_file(r"C:\Users\marcos perrude\Downloads\BR_Municipios_2022\BR_Municipios_2022.dbf")

# Converte a coluna 'CD_MUN' e 'CODIGO IBGE' para string
brasil_municipios['CD_MUN'] = brasil_municipios['CD_MUN'].astype(str)
df['CODIGO IBGE'] = df['CODIGO IBGE'].astype(str)

# Faz o merge entre o shapefile e os dados do dataframe 'df'
mapa_merged = brasil_municipios.merge(df, left_on='CD_MUN', right_on='CODIGO IBGE', how='left')

# Configura uma escala logarítmica para melhor visualização da concentração
fig, ax = plt.subplots(1, 1, figsize=(15, 15))

# Normaliza os valores para a escala logarítmica (evita log(0) adicionando um pequeno valor)
norm = mcolors.LogNorm(vmin=mapa_merged['TOTAL'].min() + 1, vmax=mapa_merged['TOTAL'].max())

# Plota o mapa com uma escala de cor contínua para 'TOTAL'
mapa_merged.plot(column='TOTAL', ax=ax, legend=True, cmap='OrRd', edgecolor='black', norm=norm)

# Título e exibição do mapa
plt.title('Mapa de Concentração do TOTAL por Município do Brasil (Escala Logarítmica)')
plt.show()
```
